##### ESERCIZIO 1
Siano date due variabili aleatorie indipendenti $X$ e $Y$ che seguono una distribuzione normale dove $\mu_X$ è il valore atteso di $X$ e $\mu_Y$ il valore atteso di $Y$, e dove $\sigma_X$ è la deviazione standard di $X$ e $\sigma_Y$ la deviazione standard di $Y$.
Sia data $Z = X - Y$, rispondere alle seguenti domande:
1. $Z$ è una variabile aleatoria? Se sì, è discreta o continua? Motivare adeguatamente la risposta

**Soluzione**

La variabile aleatoria $Z = X - Y$ è sicuramente continua in quanto differenza di variabili aleatorie continue

2. Calcolare il valore atteso di $Z$ esprimendolo in funzione di $\mu_X$ e $\mu_Y$, mostrando i passaggi

**Soluzione**

Il valore atteso di $Z$ è dato da: $E[Z] = E[X - Y] = E[X] - E[Y] = \mu_X - \mu_Y$

3. Calcolare la varianza di $Z$ esprimendola in funzione di $\sigma_X$ e $\sigma_Y$, mostrando i passaggi

**Soluzione**

Per calcolare la varianza andiamo ad esprimerla allo stesso modo in cui abbiamo fatto il calcolo precedente: 

$ Var(Z) = Var(X - Y) = Var(X) + Var(-Y) = \sigma_X^2 + \sigma_Y^2$

4. La variabile aleatoria $Z$ segue una distribuzione normale? Motivare la risposta

**Soluzione**

Beh si che segue una distribuzione normale, proviamo a pensarla nella visione del grafico, la normale segue una distribuzione determinata da una gaussiana, o normale, se io vado a sottrarre i dati delle due gaussiane otterrò una gaussiana scalata teoricamente. E' come se andassi a scalare la gaussiana X di una costante Y

5. Calcolare la funzione di ripartizione $F_Z(x)$ esprimendola in funzione dei parametri citati sopra e utilizzando $\phi$

**Soluzione**

$Var(Z) = \sigma_X^2 + \sigma_Y^2$\
$\sigma_Z = \sqrt{\sigma_X^2 + \sigma_Y^2}$\
$E[Z] = \mu_X - \mu_Y$\
$N(\mu_Z,\sigma_Z)$

$F_Z(x) = P(Z \le x) = \phi(\frac{x - \mu_Z}{\sigma_Z}) = \phi(\frac{x - \mu_X + \mu_Y}{\sqrt{\sigma_X^2 + \sigma_Y^2}})$

6. Tenendo conto che $\mu_X = 2$, $\mu_Y = 1$, $\sigma_X = 1$ e $\sigma_Y = 2$, disegnare il grafico della funzione di ripartizione di $Z$

**Soluzione**

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

mu_x = 2
mu_y = 1
sigma_x = 1
sigma_y = 2

sigma_z = np.sqrt(1**2 + 2**2)
mu_z = mu_x - mu_y

x = np.linspace(mu_z - 3 * sigma_z, mu_z + 3 * sigma_z, 500)
y = stats.norm.cdf((x - mu_z)/sigma_z)

plt.scatter(x, y)
plt.show()
```

7. Calcolare la funzione di densità di probabilità $f_Z(x)$ esprimendola in funzione dei parametri sopracitati e utilizzando $\phi$

**Soluzione**

$f_Z(X) = \frac{1}{\sqrt{2\pi}\sigma_Z}*e^{\frac{-(x-\mu_z)^2}{2*\sigma_z^2}}$

questa è la formula normale per $f_Z$, che potremmo usare se non ci chiedesse espressamente la formula in funzione di $\phi$
siccome invece questo accade dobbiamo andare a fare la derivata della funzione precedente, $F_z(x)$ ovvero: 

$\frac{d}{dz}(\phi(\frac{x \mu_Z}{\sigma_Z})) = \frac{1}{\sigma_z} * \phi(\frac{x - \mu_z}{\sigma_z}) $

8. Tenendo conto che $\mu_X = 2$, $\mu_Y = 1$, $\sigma_X = 2$ e $\sigma_Y = 2$, disegnare il grafico della funzione di densità di probabilità di $Z$

**Soluzione**

```python
x = np.linspace(mu_z - 3 * sigma_z, mu_z + 3 * sigma_z, 500)
y = 1/sigma_z * (stats.norm.cdf((x-mu_z)/sigma_z))

plt.scatter(x,y)
plt.show()
```

##### ESERCIZIO 2
Siano $Z_1, \dots, Z_n$ variabili aleatorie che descrivono un campione estratto da $Z$, distribuito come l'omonima all'esercizio precedente
1. Considerando solo in questo punto che sia noto $\mu_X$ determinare uno stimatore non distorto $U$ che stimi $\mu_Y$

**Soluzione**

Uno stimatore che stimi $\mu_Y$ è uno stimatore che dato $E[U] = \mu_Y$ per il metodo plug in. Per applicarlo andiamo ad apporre il valore atteso normale, nella distribuzione normale precedente, uguale a $\mu_Y$ andando a risolvere troveremo come dobbiamo definire $U$:

$E[U] = \mu_Y = \mu_Y + \mu_X$ 

il nostro stimatore deve essere quindi uguale a $E[-Z+\mu_X]$ così facendo se otteniamo che $U = -X + \mu_X$ e quindi che per $E[U] = E[-Z + \mu_x] = E[-Z] +\mu_x = -\mu_x + \mu_y + mu_x = \mu_y$ 

2. Considerando solo in questo punto che $\sigma_X = \sigma_Y$ determinare uno stimatore non distorto $V$ che stimi $\sigma_X^2$

**Soluzione**

Uno stimatore $V$ che stimi $\sigma_x^2$ è: 

$Var(Z) = \sqrt{\sigma_x^2+\sigma_x^2} = \sqrt{2*\sigma_x^2} = \sqrt{2} * \sigma_x$ quindi lo stimatore V per $\sigma_x$ è $\frac{Var(Z)}{\sqrt{2}}$ 

3. Determinare se lo stimatore $U$ è consistente in media quadratica

**Soluzione**

Sicuramente è uno stimatore non distorto, di conseguenza:

$MSE(U) = Var(\bar{U}) = \frac{1}{n}Var(U) = \frac{\sigma_y^2}{n}$

Portando l'$MSE(U)$ all'infinito ottengo:

$\lim_{n -> +\infty} MSE(U) = 0$ 

Quindi si è consistente in media quadratica

4. È possibile fare lo stesso ragionamento per $V$? Se no motivare la risposta, altrimenti determinare se lo stimatore è consistente in media quadratica

**Soluzione**

Si è possibile fare lo stesso ragionamento, considerando che, siccome il mio stimatore è dato in varianza, bisogna inserire la varianza della varianza, ma questo sul risultato finale non cambia niente e fa diventare comunque il tutto tendente a 0 all'infinito

5. Dato $\epsilon$, determinare la probabilità che lo stimatore $U$ presenti al più un errore di $\epsilon$
