### Esercizio 1
Considerate la funzione definita da
$$
f(x) = (m-x) \dfrac{a}{m} \, I_{[0,m]}(x),
$$
dove $a,m \in \mathbb R^+$.

1. Ricavate il valore di $a$, espresso in funzione di $m$, che rende $f$ una densità di probabilità.

    $$
    \int_{0}^{m} \frac{\alpha}{m}(m-x)\,dx = 1
    $$
    $$
    \frac{\alpha}{m}\int_{0}^{m}(m-x)\,dx=\frac{\alpha}{m}\Bigl[mx-\frac{x^{2}}{2}\Bigr]_{0}^{m}
    =\frac{\alpha}{m}\,\frac{m^{2}}{2}= \frac{\alpha m}{2}=1
    \Longrightarrow \boxed{\;\alpha=\dfrac{2}{m}\;}
    $$
    D'ora in avanti, fisseremo $\alpha$ al valore determinato al punto precedente.
    $$
    f(x)=\frac{2}{m^{2}}(m-x)\, I_{[0,m]}(x)
    $$

2. Sia $X$ una variabile aleatoria la cui distribuzione è descritta dalla densità $f$. Ricavate il valore atteso di $X$, esprimendolo in funzione di $m$.

    $$
    \begin{aligned}
    \mathbb E[X] &= \int_{0}^{m} x\,f(x)\,dx
    =\frac{2}{m^{2}}\int_{0}^{m}(mx-x^{2})\,dx \\
    &=\frac{2}{m^{2}}\left[\frac{m x^{2}}{2}-\frac{x^{3}}{3}\right]_{0}^{m}
    =\frac{2}{m^{2}}\left(\frac{m^{3}}{2}-\frac{m^{3}}{3}\right)
    =\frac{2}{m^{2}}\cdot\frac{m^{3}}{6}
    =\boxed{\;\dfrac{m}{3}\;}
    \end{aligned}
    $$

3. Ricavate la varianza e la deviazione standard di $X$, esprimendole in funzione di $m$.

    $$
    \begin{aligned}
    \mathbb E[X^{2}] &=\frac{2}{m^{2}}\int_{0}^{m}(m x^{2}-x^{3})\,dx
    =\frac{2}{m^{2}}\left[\frac{m x^{3}}{3}-\frac{x^{4}}{4}\right]_{0}^{m}\\
    &=\frac{2}{m^{2}}\left(\frac{m^{4}}{3}-\frac{m^{4}}{4}\right)
    =\frac{2}{m^{2}}\cdot\frac{m^{4}}{12}
    =\frac{m^{2}}{6}
    \end{aligned}
    $$
    $$
    \operatorname{Var}(X)=\mathbb E[X^{2}]-\bigl(\mathbb E[X]\bigr)^{2}
    =\frac{m^{2}}{6}-\left(\frac{m}{3}\right)^{2}
    =\frac{m^{2}}{6}-\frac{m^{2}}{9}
    =\boxed{\; \frac{m^{2}}{18}\;}
    $$
    $$
    \sigma_X=\sqrt{\operatorname{Var}(X)}=\boxed{\;\dfrac{m}{3\sqrt{2}}\;}
    $$

4. Ricavate la forma analitica della funzione di ripartizione di $X$, che indicheremo con $F$, esprimendola in funzione di $m$.

    $$
    \begin{aligned}
    F(x) &=\int_{0}^{x} f(t)\,dt
    =\frac{2}{m^{2}}\int_{0}^{x}(m-t)\,dt
    =\frac{2}{m^{2}}\left[mt-\frac{t^{2}}{2}\right]_{0}^{x} \\
    &=\frac{2}{m^{2}}\Bigl(mx-\tfrac{x^{2}}{2}\Bigr)
    = \boxed{\; \left( \frac{2x}{m}-\frac{x^{2}}{m^{2}} \right) I_{[0,m]} + I_{(m, +\infty)} \;}
    \end{aligned}
    $$

5. Fissato, **solo in questo punto**, $m=2$, tracciate i grafici di $f$ e di $F$.

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import scipy.stats as sp
import matplotlib.pyplot as plt

m = 2
def f(x, m):
    if x < 0 or x > m:
        return 0
    return 2/m - (2*x)/(m**2)

def F(x, m):
    if x < 0:
        return 0
    elif x > m:
        return 1
    return (2*x)/m - (x**2)/(m**2)

x = np.linspace(-1, m+1, 200)
y = [f(i, m) for i in x]
Y = [F(i, m) for i in x]

plt.plot(x, y, label='f(x)')
plt.plot(x, Y, label='F(x)')
plt.legend()
plt.xlabel('x')
plt.ylabel('Density / CDF')
plt.show()
```

6. Sia $Y$ una variabile aleatoria distribuita secondo il modello esponenziale. Ricavate il valore del suo parametro $\lambda$, esprimendolo in funzione di $m$ e in modo che risulti $\mathbb E[X] = \mathbb E[Y]$.

    $$
    Y\sim E(\lambda), \qquad \mathbb E[Y] = \dfrac{1}{\lambda}
    $$
    $$
    \mathbb E[X]=\mathbb E[Y] \quad \Rightarrow \quad \frac{1}{\lambda}=\frac{m}{3}\quad\Rightarrow\quad
    \boxed{\;\lambda=\dfrac{3}{m}\;}
    $$

7. Calcolate la varianza di $Y$, esprimendola in funzione di $m$. Calcolate poi il rapporto $\text{Var}(Y)/\text{Var}(X)$, commentando il risultato ottenuto.

    $$
    \operatorname{Var}(Y)=\frac{1}{\lambda^{2}}
    =\left(\frac{m}{3}\right)^{2}
    =\boxed{\;\dfrac{m^{2}}{9}\;}
    $$

    $$
    \operatorname{Var}(X)=\frac{m^{2}}{18}.
    $$

    $$
    \frac{\operatorname{Var}(Y)}{\operatorname{Var}(X)}
    =\frac{m^{2}/9}{m^{2}/18}
    =\frac{1}{9}\cdot\frac{18}{1}
    =\boxed{\;2\;}
    $$

    Nonostante $X$ e $Y$ presentino lo stesso valore atteso e quindi la stessa centralità, la distribuzione esponenziale $Y$ è due volte più dispersa di quella con densita lineare di $X$.

### Esercizio 2
Siano $X_1, \dots, X_n$ delle variabili aleatorie che modellano un campione estratto da una popolazione $X$ distribuita come l'omonima variabile aleatoria dell'esercizio precedente, per un valore ignoto del parametro $m$.

1. Determinate uno stimatore non distorto $T$ per $m$.

    $$
    \mathbb E[\bar X]=\mathbb E[X]=m/3
    $$
    $$
    \boxed{\,T = 3\bar X\;} \qquad  \mathbb E[T]=\mathbb 3E[\bar X] = 3\, \dfrac{m}{3} = m
    $$

2. Determinate uno stimatore non distorto $U$ per la deviazione standard di $X$.

    $$
    \boxed{\,U=\frac{\bar X}{\sqrt2}\;} \qquad
    \mathbb E[U]=\frac{\mathbb E[\bar X]}{\sqrt2}
    =\frac{m/3}{\sqrt2}
    =\frac{m}{3\sqrt2}=\sigma_X
    $$

3. Determinate se gli stimatori $T$ e $U$ che avete ottenuto al punto precedente godono della proprietà di consistenza quadratica rispetto alla quantità che stimano.

    $$
    MSE_{m}(T) = \operatorname{Var}(T)=9\operatorname{Var}(\bar X)
    =9\frac{\operatorname{Var}(X)}{n}
    =9\frac{m^{2}/18}{n}
    =\frac{m^{2}}{2n} \quad \xrightarrow[n\to\infty]{}0
    $$
    $$
    MSE_{\sigma_X}(U) = \operatorname{Var}(U)=\frac{1}{2}\operatorname{Var}(\bar X)
    =\frac{1}{2}\frac{m^{2}/18}{n}
    =\frac{m^{2}}{36n} \quad \xrightarrow[n\to\infty]{}0
    $$
    Entrambi gli stimatori sono quindi consistenti in media quadratica.

4. Indicato con $\varepsilon$ un valore reale positivo, calcolate le seguenti probabilità:

    - $\mathbb P(|T - m| \le \varepsilon)$,
    - $\mathbb P(|U - \sigma_X| \le \varepsilon)$.

    $$
    T\;\dot\sim\; N \Bigl(m,\;\underbrace{\frac{m^{2}}{2n}}_{\sigma_{T}^{2}}\Bigr)
    \qquad
    \sigma_T=\frac{m}{\sqrt{2n}}
    $$

    $$
    \boxed{
    \mathbb P (|T-m|\le\varepsilon)\;\approx\;2\Phi\!\Bigl(\frac{\varepsilon}{\sigma_T}\Bigr)-1
    \;=\;2\Phi\!\Bigl(\frac{\varepsilon\sqrt{2n}}{m}\Bigr)-1 }
    $$

    $$
    U\;\dot\sim\;\mathcal N \Bigl(\sigma,\;\underbrace{\frac{m^{2}}{36n}}_{\sigma_{U}^{2}}\Bigr)
    \qquad
    \sigma_U=\frac{m}{6\sqrt n}
    $$

    $$
    \boxed{
    \mathbb P (|U-\sigma|\le\varepsilon)\;\approx\;2\Phi\!\Bigl(\frac{\varepsilon}{\sigma_U}\Bigr)-1
    \;=\;2\Phi\!\Bigl(\frac{6\varepsilon\sqrt n}{m}\Bigr)-1 }
    $$

5. Fissato $\delta \in \mathbb R^+$, indichiamo:

    - con $n_1$ la più piccola dimensione del campione che garantisce che $\mathbb P(|T - m| \le \varepsilon) \ge 1 - \delta$,

    $n_1 > n_2$ in quanto presentano lo stesso numeratore e il numeratore di $n_2$ è più grande.
