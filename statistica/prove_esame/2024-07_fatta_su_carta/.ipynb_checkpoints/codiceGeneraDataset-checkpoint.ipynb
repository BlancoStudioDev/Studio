```python
import numpy as np
import pandas as pd

n = 1000  # numero di campioni

# Parametri
a = 40 # massimo punteggio maggiore di 36
x = np.arange(1, a+1)

# Probabilità secondo la legge
prob = 2*x / (a*(a+1))
prob = prob / prob.sum()  # normalizzazione

# Calcolo numero di campioni per ogni punteggio
frequenze = np.round(prob * n).astype(int)

# Aggiustiamo la somma per essere esattamente n
diff = n - frequenze.sum()
if diff > 0:
    frequenze[-diff:] += 1
elif diff < 0:
    frequenze[:abs(diff)] -= 1

# Creiamo l'array ordinato dei punteggi
punteggio = np.hstack([np.full(f, val) for f, val in zip(frequenze, x)])

# Aggiungiamo piccola componente continua (0-1) senza superare 'a'
punteggio = punteggio + np.random.uniform(0,1,len(punteggio))
punteggio = np.clip(punteggio, 1, a)

# Ordiniamo per avere punteggio crescente
punteggio.sort()

# Altri attributi
matricola = np.arange(1, n+1)
genere = np.random.choice(['M','F'], size=n)
eta = np.random.randint(18, 60, size=n)
tempo = np.random.randint(10,121, size=n)

# Creiamo DataFrame
df = pd.DataFrame({
    'matricola': matricola,
    'genere': genere,
    'eta': eta,
    'punteggio': punteggio,
    'tempo': tempo
})

# Salvataggio CSV
df.to_csv('dataset_crescente.csv', index=False)

# Verifica
print(df.head())
print("\nCorrelazione punteggio-eta:", df['punteggio'].corr(df['eta']))
```
