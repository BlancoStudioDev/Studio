1. Quali sono le specificazioni di Z?

```python
# X = 0, Y = 0: Z = 0
# X = 1, Y = 0: Z = 2
# X = 0, Y = 1: Z = -1
# X = 1, Y = 1: Z = 1
```

2. Calcolate P(Z = z) per tutte le specificazioni che avete determinato nel punto precedente

```python
# Usa la probabilità congiunta quindi X.pmf * Y.pmf

# P(Z)=0 -> (1-p)(1-p)
# P(Z)=2 -> p(1-p)
# P(Z)=-1 -> p(1-p)
# P(Z)=1 -> p**2
```

3. Solo in questo punto, assumiamo che p=3/4. Visualizzate graficamente la distribuzione Z (potete farlo sia a mano, sia in codice) giustificando il tipo di grafico scelto

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st

p=3/4

z_range = [-1,0,1,2]
z_probs = [p*(1-p), (1-p)*(1-p), p**2, p*(1-p)]

plt.vlines(z_range, 0, z_probs)
plt.show()

# ho scelto un vlines per mostrare quanto probabili sono i singoli outcomes in relazione fra di loro
```

4. Calcolate il valore atteso e la varianza di Z, esprimendoli in funzione di p e giustificando i passaggi svolti  

```python
# E(Z) = 2*E(X)-E(Y) = 2*p - p = p
# Var(Z) = 2*Var(X) - Var(Y) = 4 p*(1-p) + p(1-p) = 5p(1-p)
```

5. Mostrate un grafico che illustra come cambia Var(Z) al variare di p. Determinate poi un limite superiore per Var(Z) che non dipende da p

```python
p_range = np.linspace(0,1)
var_range = [5*p*(1-p) for p in p_range]

plt.plot(p_range, var_range)
plt.show()

5*1/2*1/2 # dato che è una parabola, il limite superiore sta al centro p=1/2 -> 1.25
```

Consideriamo una popolazione distribuita come una v.a. Z, la cui distribuzione è la stessa dell'omonima variabile definita nell'esercizio precedente, per un valore ignoto del parametro p. per n € Naturali fissato, siano Z1,...,Zn delle v.a. che descrivono un campione estratto da questa popolazione. In tutto l'esercizio Z' indicherà la media campionaria calcolata su Z1,...,Zn.


1. Z' è uno stimatore non deviato per il valore atteso di Z? 

```python
# Sì, Z' è non deviato perché la media campionaria è SEMPRE uno stimatore non distorto per il valore atteso del suo campione
```

2. Z' è uno stimatore non deviato per p?

```python
# Sì è non deviato perché E(Z)=p e nell'es di prima abbiamo provato che Z' è non deviato per E(Z)
```

3. Giustificando i passaggi, calcolate lo scarto quadratico medio di Z' nella stima di p, esprimendolo in funzione di n e p

```python
# MSEz'p = Var(Z')+bias**2 = Var(Z)/n + (E(Z')-p)**2 = 5*p*(1-p)/n + 0  
```

4. Sia sigma' la deviazione standard di Z', ricavate un limite superiore per sigma' che dipenda solo dalla dimensione del campione

```python
n_range = np.arange(1,100)

p=0.5

def sigma_medio(n):
    return np.sqrt(5*p*(1-p)/n)

y_range = [sigma_medio(n) for n in n_range]

plt.plot(n_range, y_range)
plt.show()

# il limite superiore è la media di 1 sola v.a.
print(f"limite superiore di sigma': {sigma_medio(1)}")
```

5. Fissato epsilon>0, applicando il TCL, esprimete in funzione di epsilon, n e sigma' la probabilità dell'evento che si verifica quando l'errore in valore assoluto che si compie usando Z' per stimare p è minore u uguale di epsilon

```python
# P(|Z'-p| <= epsilon) -> 2*phi(epsilon*sqrt(n)/sqrt(sigma')) - 1
def duephimenouno(epsilon, sigma_medio):
    return 2*st.norm.cdf(epsilon/sigma_medio) - 1

# tutto il calcolo per ripasso
# P(-epsilon <= Z'-p <= epsilon)
# P(Z'-p>=-epsilon) + P(Z'-p <= epsilon)
# W = Z'-p/sigma'
# P(W>= -epsilon/sigma') + P(W <= epsilon/sigma')
# phi(epsilon/sigma') + (1 - P(W <= epsilon/sigma'))
# 2*phi(epsilon/sigma') - 1
# sigma' = sigma/sqrt(n)
# 2*phi(epsilon*sqrt(n)/sigma)
```

6. Calcolate un limite inferiore per la probabilità calcolata nel punto precedente, che dipenda solo da epsilon e n

```python
x_range = np.arange(1, 33)
y_range = [duephimenouno(1, sigma_medio(n)) for n in x_range]

plt.plot(x_range, y_range)
plt.show()

# non so se è giusto o sbagliato
print(f"limite inferiore: {duephimenouno(1, sigma_medio(1))}")
```

2. Visualizzate la distribuzione dell’attributo tipo_motore, motivando la scelta dello strumento grafico utilizzato.

```python
auto['tipo_motore_enum'] = auto['tipo_motore'].astype('category').cat.codes

auto.tipo_motore_enum.plot.hist(bins=20)
plt.show()
```

3. Utilizzando un indice numerico, valutate quanto i possibili differenti valori per l’attributo tipo_motore siano equamente distribuiti nel dataset, motivando l’indice utilizzato e discutendo i risultati ottenuti, confrontandoli con il grafico ottenuto al punto precedente.

```python
freq = auto.tipo_motore_enum.value_counts(normalize=True)

# puoi usare entropia
def entropy(freq):
    # Normalized entropy (Shannon)
    return -np.sum(freq * np.log2(freq)) / np.log2(len(freq))

# o gini
def gini(freq):
    # Normalized Gini index
    return 1 - np.sum(freq**2) / (len(freq)-1)/len(freq)

print("entropia: ", entropy(freq))
print("gini: ", gini(freq))
```

```python
# 4. usando la stima di p, calcola le seguenti probabilità:
# P(X = 0)
print(f"P(X=0) = {(1-p)*(1-p)}")
# P(X <= 0)
print(f"P(X<=0) = P(X=0) + P(X=-1) = {(1-p)*(1-p)+p*(1-p)}")
# P(X pari)
print(f"P(X=0) + P(X=2) = {(1-p)*(1-p)+p*(1-p)}")
```

```python
# 5. c'è una tariffa annua uguale a 100+30z dove z indica il punteggio.
# calcola T, il suo valore atteso, la sua varianza e la prob. T < 100

auto['tariffa'] = 100+30*auto['punteggio']

# E(T) = 100+30*E(Z) = 100+30p
mu = 100+30*p
# Var(T) = 30^2 * Var(Z) = 900*5p(1-p) codue c'è qualcosa che non va
var = 900*5*p*(1-p)
# P(Tariffa < 100) puoi magheggiare e vedere che tariffa < 100 quando P(Z=-1)
probt = p*(1-p)


print("mu: ",mu, ", var: ",var, "P(T<100): ",probt)
```

```python
# 6. Il comune stima che l'anno prossimo 3000 veicoli pagheranno la tariffa. 
# Indica con G la v.a. che indica il totale degli introiti che incasserà il comune.
# Mostra o approssima la distribuzione di G

# G = 3000*T quindi può essere approssimato da una normale
# G è approssimato da N(100+30p*n, sqrt(900*5*p*(1-p)*n))

n = 3000
G = st.norm(mu*n, np.sqrt(var*n))

x = np.linspace(G.mean() - 4*G.std(), G.mean() + 4*G.std(), 1000)
plt.plot(x, G.pdf(x))
plt.show()
```

```python
# 7. la probabilità che G > 310000 
# 1 - P(G<=310000)
1 - G.cdf(310000)
```
